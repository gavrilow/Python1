## Calculate the area of triangle with sides: a, b, c

```python
from math import pi, sqrt
```

```python
a = 4.5
b = 5.9
c = 9
# Heron's Formula
s = (a + b + c) / 2 # half of the triangles perimeter
area = sqrt(s * (s - a) * (s - b) * (s - c)) # calculate the area
area = round(area, 2)
print (f'The area of a triangle with sides: {a}, {b}, {c} is {area}')
```

## Build a graph of functions sin(x) and cos(x) on the interval (-4pi; 4pi)

```python
from math import pi, sqrt, sin, cos
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
#Graphs are built into the notebook
```

```python
x = np.linspace(- 4 * pi, 4 * pi, 100)

#plot size in inches
plt.figure(figsize=(13, 6))

# Set the functions: 
plt.plot(x,np.sin(x), color='red', marker='.', linestyle='--', label='sin(x)')
plt.plot(x,np.cos(x), color='blue', marker='*', label='cos(x)')
plt.legend(loc='upper center', ncol=2)
plt.grid()

#Signatures for axes: 
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

#Y-axis range: 
plt.ylim(-1.1, 1.4)

#Title: 
plt.title(r'Functions graph $y = sin (x) $  $y = cos(x)$', fontsize=16, y=1.05);
```

## Build a graph of function sin(sqrt(x^2 + y^2))

```python
from math import pi, sqrt, sin, cos
import numpy as np
import matplotlib.pyplot as plt

#Graphs are built into the notebook
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(- 4 * pi, 4 * pi, 50)
y = np.linspace(- 4 * pi, 4 * pi, 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.figure(figsize=(13, 13))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')

ax.set_title('sin(sqrt(x^2 + y^2))');
ax.view_init(60, 25)
```
